```python
#import the csv module
import csv
```

```python
#store the csv file into variable
countries_list = 'Countries_list.csv'
```

```python
#initialize the columns and rows lists
columns = []
rows = []
```

```python
#read the csv file content
with open(countries_list, 'r') as csvfile:
  csvreader = csv.reader (csvfile)

  #extract the columns in the first row
  columns = next(csvreader)

  #extract each data into rows
  for row in csvreader:
    rows.append(row)

```

# Guess a random country based on the provided list above

*   Hints will be provided - describing if the guessed country's population is bigger or smaller than the correct answer.
*   Only five guesses will be allowed until the game is over.

*Good luck!* 





---



```python
#import the necessary module
import random
```

```python
#defined function to initiate the game

def guess_a_country_game():
  try:
    random_country = random.choice(csvreader_list)

    #show the generated country - for testing purposes only
    print(random_country[1])

    #initialize the counter (10 for maximum ten guesses only before the game is over)
    counter = 5

    #loop to accept a guess and check every input. Hints will be provided accdg to land area differences
    for counter in range (5,0,-1):
      guess = input('Guess a country: ')



      #compare the guessed country and the generated random country
      if guess == random_country[1]:
        print('Correct! Congratulations! Lucky guess! haha')
        print(f'{random_country[1]} has a population of {random_country[2]} (as of 2023) and it is the ranked as the number {random_country[0]} among 196 countries in terms of population (descending order).')
        break
      else:
        print(f'Wrong! Try again. Remaining attempts: {counter - 1}')

        #reduce the counter due to one used attempt
        counter -= 1

        #extract the information of this current guess
        guess_rank = []
        for i in range (0, csvreader.line_num):
          if str(guess) in csvreader_list[i]:
            guess_rank.insert(0, csvreader_list[i])
            break
          i+=1

        #give a hint by giving perspective on the size
        print(f'Your current guess "{guess_rank[0][1]}" has a population of {guess_rank[0][2]} (as of 2023) and it is the ranked as number {guess_rank[0][0]} among 196 countries in terms of population (descending order).')

        int_guess_rank = int(guess_rank[0][0])
        # print(int_guess_rank)
        # print(type(int_guess_rank))
        int_random_country= int(random_country[0])
        # print(int_random_country_hint)
        # print(type(int_random_country_hint))
        if int_guess_rank > int_random_country :
          print(f"The correct country's population is bigger than this guess.")
        elif int_guess_rank < int_random_country :
          print(f"The correct country's population is smaller than this guess.")
        else:
          print('invalid answer')
          break


    if counter == 0:
      print(f'You have used up all the attempts. The correct answer is: {random_country[1]}')
      print(f'{random_country[1]} has a population of {random_country[2]} (as of 2023) and it is the ranked as the number {random_country[0]} among 196 countries in terms of population (descending order).')


  except:
    print('Invalid answer.')
```
